```python
import matplotlib.pyplot as plt
import wfdb
import numpy as np
import pickle
import pywt
from sklearn import preprocessing
from scipy import io
from scipy.signal import resample
```

```python
def z_score(A, lead):#输入心电图维度(lead,sample)
    mean=np.sum(A,axis=-1).reshape(lead,1)/len(A[0])
    B=(A-mean)**2
    V=np.sum(B,axis=-1).reshape(lead,1)/len(B[0])
    S=V**(1/2)
    A=(A-mean)/S
    return A
```

```python
c_t = 0.15#前向截取长度对应时间
fz = 360#采样频率
re_sample = 256#重采样心拍长度

def data_process1(i,num,begin):
    heartbeat_signal = record_wavelet[:,signal_annotation.sample[i]-int(RR_mean[i]-c_t*fz)-begin:signal_annotation.sample[i]+int(2/3*RR_mean[i])-begin]
    heartbeat_signal_resample = resample(heartbeat_signal,num,axis=1)
    heartbeat_signal_resample = np.array(heartbeat_signal_resample)
    heartbeat_signal_resample = z_score(heartbeat_signal_resample,2)
    return heartbeat_signal_resample

def data_append(i,class_num,t):
    a=0
    while a<t:
        Train_ecg_data.append(data_process1(i,re_sample,0))
        Train_label.append([class_num])
        RR_Train.append(RR_infor[:,i])
        a+=1
        if a>=t:
            break
```

```python
#Random训练集
N_label = ['N','L','R','e','j']
S_label = ['A','a','J','S']
V_label = ['V','E']
F_label = ['F']
Q_label = ['Q']

Train_Records = [101,106,108,109,112,114,115,116,118,119,122,124,201,203,205,207,208,209,215,220,223,230, 100, 103, 105, 111, 113, 117, 121, 123, 200, 202, 210, 212, 213, 214, 219, 221, 222, 228, 231, 232, 233, 234]

Train_ecg_data = []
Train_label = []
RR_Train = []

for m in range(len(Train_Records)):
    record = wfdb.rdrecord('/home/wj/Downloads/mit-bih-arrhythmia-database-1.0.0/' + str(Train_Records[m]),physical=True,sampfrom=0,sampto=None,channels=[0,1])
    signal_annotation = wfdb.rdann('/home/wj/Downloads/mit-bih-arrhythmia-database-1.0.0/' + str(Train_Records[m]),'atr',sampfrom=0,sampto=None)
    
    #手工特征
    RR_infor = []
    RR_index = signal_annotation.sample
    RR = np.array(RR_index-([0]+list(RR_index))[:-1])#得到RR间期数组，第三个值开始有意义，即索引为2
    RR_diff1 = np.array(RR-([0]+list(RR))[:-1])#得到RR间期的一阶差分,索引为3代表第3个R波前后的间期差分
    
    RR_diff1 = np.array((list(RR_diff1)+[0])[1:])#将diff1变成于RR对应的索引
    RR_nor=(RR-np.min(RR[3:-3]))/(np.max(RR[3:-3])-np.min(RR[3:-3]))
    RR_diff1_nor = (RR_diff1-np.min(RR_diff1[3:-3]))/(np.max(RR_diff1[3:-3])-np.min(RR_diff1[3:-3]))
    RR_infor.append(RR_nor)
    RR_infor.append(RR_diff1_nor)
    RR_infor = np.array(RR_infor)
    
    RR_expand = [0]*5+list(RR)+[0]*6
    RR_mean = []
    for k in range(len(RR)-1):
        RR_part_mean = (np.sum(RR_expand[k:k+11])-np.min(RR_expand[k:k+11]))/10
        RR_mean.append(RR_part_mean)
    RR_mean = np.array(RR_mean)
        
    record_wavelet = []
    for n in range(len(record.p_signal[1])):
        coeffs = pywt.wavedec(record.p_signal[:,n], 'coif4', level=8)
        #置0第10尺度近似分量
        coeffs[-9:-8] = np.zeros_like(coeffs[-9:-8])
        coeffs[-3:-2] = np.zeros_like(coeffs[-3:-2])
        coeffs[-2:-1] = np.zeros_like(coeffs[-2:-1])
        coeffs[-1:] = np.zeros_like(coeffs[-1:])
        ECG_new2 = pywt.waverec(coeffs, 'coif4')
        ECG_new2 = ECG_new2.flatten()
        record_wavelet.append(ECG_new2)
    record_wavelet = np.array(record_wavelet)

    for i in range(2,len(signal_annotation.sample)-1):
        
        if signal_annotation.symbol[i] == N_label[0]:
            Train_ecg_data.append(data_process1(i,re_sample,0))
            Train_label.append([0])
            RR_Train.append(RR_infor[:,i])
            data_append(i,0,0)
        if signal_annotation.symbol[i] == N_label[1]:
            Train_ecg_data.append(data_process1(i,re_sample,0))
            Train_label.append([0])
            RR_Train.append(RR_infor[:,i])
            data_append(i,0,0)
        if signal_annotation.symbol[i] == N_label[2]:
            Train_ecg_data.append(data_process1(i,re_sample,0))
            Train_label.append([0])
            RR_Train.append(RR_infor[:,i])
            data_append(i,0,0)
        if signal_annotation.symbol[i] == N_label[3]:
            Train_ecg_data.append(data_process1(i,re_sample,0))
            Train_label.append([0])
            RR_Train.append(RR_infor[:,i])
        if signal_annotation.symbol[i] == N_label[4]:
            Train_ecg_data.append(data_process1(i,re_sample,0))
            Train_label.append([0])
            RR_Train.append(RR_infor[:,i])
        if signal_annotation.symbol[i] == S_label[0]:            
            Train_ecg_data.append(data_process1(i,re_sample,0))
            Train_label.append([1])
            RR_Train.append(RR_infor[:,i])
            data_append(i,1,0)
        if signal_annotation.symbol[i] == S_label[1]:            
            Train_ecg_data.append(data_process1(i,re_sample,0))
            Train_label.append([1])
            RR_Train.append(RR_infor[:,i])
            data_append(i,1,0)
        if signal_annotation.symbol[i] == S_label[2]:            
            Train_ecg_data.append(data_process1(i,re_sample,0))
            Train_label.append([1])
            RR_Train.append(RR_infor[:,i])
            data_append(i,1,0)
        if signal_annotation.symbol[i] == S_label[3]:            
            Train_ecg_data.append(data_process1(i,re_sample,0))
            Train_label.append([1])
            RR_Train.append(RR_infor[:,i])
        if signal_annotation.symbol[i] == V_label[0]:
            Train_ecg_data.append(data_process1(i,re_sample,0))
            Train_label.append([2])
            RR_Train.append(RR_infor[:,i])
            data_append(i,2,0)
        if signal_annotation.symbol[i] == V_label[1]:
            Train_ecg_data.append(data_process1(i,re_sample,0))
            Train_label.append([2])
            RR_Train.append(RR_infor[:,i])
            data_append(i,2,0)
        if signal_annotation.symbol[i] in F_label:
            Train_ecg_data.append(data_process1(i,re_sample,0))
            Train_label.append([3])
            RR_Train.append(RR_infor[:,i])
            data_append(i,3,0)

            
Train_ecg_data = np.array(Train_ecg_data)
Train_label = np.array(Train_label)
RR_Train = np.array(RR_Train)

print(Train_ecg_data.shape)
print(Train_label.shape)
print(RR_Train.shape)
```

```python
#DS2测试集去噪
N_label = ['N','L','R','e','j'] 
S_label = ['A','a','J','S'] 
V_label = ['V','E'] 
F_label = ['F'] 
Q_label = ['Q'] 
Test_Records= [100, 103, 105, 111, 113, 117, 121, 123, 200, 202, 210, 212, 213, 214, 219, 221, 222, 228, 231, 232, 233, 234]

Test_ecg_data = [] 
Test_label = []
RR_Test = []

sample_from = 108000
sample_to = 650000

for g in range(len(Test_Records)):
    record = wfdb.rdrecord('/home/wj/Downloads/mit-bih-arrhythmia-database-1.0.0/' + str(Test_Records[g]),physical=True,sampfrom=sample_from,sampto=sample_to,channels=[0,1])
    signal_annotation = wfdb.rdann('/home/wj/Downloads/mit-bih-arrhythmia-database-1.0.0/' + str(Test_Records[g]),'atr',sampfrom=sample_from,sampto=sample_to)
    
    #手工特征    
    RR_infor = []
    RR_index = signal_annotation.sample-sample_from
    RR = np.array(RR_index-([0]+list(RR_index))[:-1])#得到RR间期数组，第三个值开始有意义，即索引为2
    RR_diff1 = np.array(RR-([0]+list(RR))[:-1])#得到RR间期的一阶差分,索引为3代表第3个R波前后的间期差分
    
    RR_diff1 = np.array((list(RR_diff1)+[0])[1:])#将diff1变成于RR对应的索引
    RR_nor=(RR-np.min(RR[3:-3]))/(np.max(RR[3:-3])-np.min(RR[3:-3]))
    RR_diff1_nor = (RR_diff1-np.min(RR_diff1[3:-3]))/(np.max(RR_diff1[3:-3])-np.min(RR_diff1[3:-3]))
    RR_infor.append(RR_nor)
    RR_infor.append(RR_diff1_nor)
    RR_infor = np.array(RR_infor)
    
    RR_expand = [0]*5+list(RR)+[0]*6
    RR_mean = []
    for k in range(len(RR)-1):
        RR_part_mean = (np.sum(RR_expand[k:k+11])-np.min(RR_expand[k:k+11]))/10
        RR_mean.append(RR_part_mean)
    RR_mean = np.array(RR_mean)
    
    record_wavelet = []
    for n in range(len(record.p_signal[1])):
        coeffs = pywt.wavedec(record.p_signal[:,n], 'coif4', level=8)
        #置0第10尺度近似分量
        coeffs[-9:-8] = np.zeros_like(coeffs[-9:-8])
        coeffs[-3:-2] = np.zeros_like(coeffs[-3:-2])
        coeffs[-2:-1] = np.zeros_like(coeffs[-2:-1])
        coeffs[-1:] = np.zeros_like(coeffs[-1:])
        ECG_new2 = pywt.waverec(coeffs, 'coif4')
        ECG_new2 = ECG_new2.flatten()
        record_wavelet.append(ECG_new2)
    record_wavelet = np.array(record_wavelet)

    for i in range(2,len(signal_annotation.sample)-1):
        
        if signal_annotation.symbol[i] == N_label[0]:
            Test_ecg_data.append(data_process1(i,re_sample,108000))
            Test_label.append([0])
            RR_Test.append(RR_infor[:,i])
        if signal_annotation.symbol[i] == N_label[1]:
            Test_ecg_data.append(data_process1(i,re_sample,108000))
            Test_label.append([0])
            RR_Test.append(RR_infor[:,i])
        if signal_annotation.symbol[i] == N_label[2]:
            Test_ecg_data.append(data_process1(i,re_sample,108000))
            Test_label.append([0])
            RR_Test.append(RR_infor[:,i])
        if signal_annotation.symbol[i] == N_label[3]:
            Test_ecg_data.append(data_process1(i,re_sample,108000))
            Test_label.append([0])
            RR_Test.append(RR_infor[:,i])
        if signal_annotation.symbol[i] == N_label[4]:
            Test_ecg_data.append(data_process1(i,re_sample,108000))
            Test_label.append([0])
            RR_Test.append(RR_infor[:,i])
        if signal_annotation.symbol[i] == S_label[0]:            
            Test_ecg_data.append(data_process1(i,re_sample,108000))
            Test_label.append([1])
            RR_Test.append(RR_infor[:,i])
        if signal_annotation.symbol[i] == S_label[1]:            
            Test_ecg_data.append(data_process1(i,re_sample,108000))
            Test_label.append([1])
            RR_Test.append(RR_infor[:,i])
        if signal_annotation.symbol[i] == S_label[2]:            
            Test_ecg_data.append(data_process1(i,re_sample,108000))
            Test_label.append([1])
            RR_Test.append(RR_infor[:,i])
        if signal_annotation.symbol[i] == S_label[3]:            
            Test_ecg_data.append(data_process1(i,re_sample,108000))
            Test_label.append([1])
            RR_Test.append(RR_infor[:,i])
        if signal_annotation.symbol[i] == V_label[0]:
            Test_ecg_data.append(data_process1(i,re_sample,108000))
            Test_label.append([2])
            RR_Test.append(RR_infor[:,i])
        if signal_annotation.symbol[i] == V_label[1]:
            Test_ecg_data.append(data_process1(i,re_sample,108000))
            Test_label.append([2])
            RR_Test.append(RR_infor[:,i])
        if signal_annotation.symbol[i] in F_label:
            Test_ecg_data.append(data_process1(i,re_sample,108000))
            Test_label.append([3])
            RR_Test.append(RR_infor[:,i])
            
Test_ecg_data = np.array(Test_ecg_data) 
Test_label = np.array(Test_label)
RR_Test = np.array(RR_Test)

print(Test_ecg_data.shape)
print(Test_label.shape)
print(RR_Test.shape)
```
